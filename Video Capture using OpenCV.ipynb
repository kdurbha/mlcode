```python
# Example from Udemy Python course on video capture using OpenCV
import cv2, time

video=cv2.VideoCapture(0)

# a is number of frames captured
a=0

#Create a while loop to capture the frames till q is pressed to quit.
print("Press q to quit")

while True:
    a= a+1
    check, frame= video.read()
    print(check)
    print(frame)

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    #time.sleep(3)
    cv2.imshow("Capturing", gray)
    key=cv2.waitKey(1)
    
    if key==ord('q'):
        break

print("Press q to quit")
print(a)

#Release the webcam
video.release()

#Close the frame window! 
#Unfortunately this command fails on Mac Jupyter Notebook! Please terminate the kernel till bug is fixed!
cv2.destroyWindow("Capturing")
cv2.waitKey(1)
```
